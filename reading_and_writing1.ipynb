```python
# Write the image

image_new = image + 100
cv2.imwrite("./output.jpg", image_new)

# Display the image
cv2_imshow(image) # Use cv2_imshow instead of cv2.imshow
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Increase the brightness of the image

    image_new = image + 30

This line creates a new image called image_new by adding 30 to each pixel value of the original image.

In the context of image processing, this operation is known as brightness enhancement.


---
---

##Reasoning:

Images are represented as arrays of pixel values.

Pixel values typically range from 0 to 255, representing intensity levels (0 being black, 255 being white).

Adding 30 to each pixel increases its intensity, making the overall image brighter.

Note that if the result of adding 30 exceeds 255, it will be capped at 255.
cv2.imwrite("./output.jpg", image_new)

This line uses the imwrite function from the OpenCV library (cv2) to save the modified image image_new to a file.

##BGR Format:

OpenCV, by default, loads images in BGR format, unlike other libraries that use RGB.

##Broadcasting:
When you perform image + 30, NumPy (the underlying library used by OpenCV) applies this operation element-wise.

This means 30 is added to each value in the image array.

3-Channel Image: Since the image is in BGR format and is represented as a 3-channel array, adding 30 affects each channel individually.
